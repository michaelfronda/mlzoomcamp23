```python
import os
import shutil
import scipy

import tensorflow
import numpy as np
import matplotlib.pyplot as plt

from tensorflow import keras

from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras import optimizers

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
```

```python
tensorflow.__version__
```

```python
# Load an image
img = image.load_img('data/train/wasp/208072345_3ce99e4e1a_n.jpg', target_size=(900, 900))
img

# ah! a giant wasp 
```

```python
# init model
cnn = models.Sequential() 

cnn.add(layers.Conv2D( # Convolutional layer 
    32, # 32 filters 
    (3,3), # kernel size 
    activation='relu', # activation 
    input_shape=(150,150,3) # shape for input, 3 channels 
))

cnn.add(layers.MaxPooling2D(2,2)) # add pooling with size (2,2)
cnn.add(layers.Flatten()) # flatten -> vector 
cnn.add(layers.Dense(64, activation='relu'))
cnn.add(layers.Dense(1, activation='sigmoid')) # ouput

cnn.compile(loss='binary_crossentropy',
            optimizer=optimizers.SGD(learning_rate=0.002, momentum=0.8), # Gradient Descent 
            metrics=['acc']) 
```

```python
train_dgen = ImageDataGenerator(rescale=1./255)
test_dgen = ImageDataGenerator(rescale=1./255)
```

```python
train_ds = train_dgen.flow_from_directory(directory = './data/train/',
                                         target_size=(150, 150), 
                                         batch_size=20, # set
                                         class_mode='binary',
                                         shuffle=True) # set


test_ds = test_dgen.flow_from_directory(directory = './data/test/',
                                     target_size=(150, 150),
                                     batch_size=20, # set
                                     class_mode='binary',
                                     shuffle=True) # set
```

```python
model = cnn.fit(
    train_ds,
    epochs=10, # set to 10; fitting via CPU for now
    validation_data=test_ds
)
```

```python
model.history
```

```python
np.median(model.history['acc']) # median accuracy 
```

```python
np.std(model.history['loss']) # std dev of loss 
```

### Augment Data Images 

```python
train_generator = ImageDataGenerator(rescale=1./255,
    rotation_range=50,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest')

test_generator = ImageDataGenerator(rescale=1./255)


train_ds = train_generator.flow_from_directory(
    './data/train',
    target_size=(150, 150),
    shuffle=True,
    batch_size=20,
    class_mode = 'binary'
)

test_ds = test_generator.flow_from_directory(
    './data/test',
    target_size=(150, 150),
    shuffle=True,
    batch_size=20,
    class_mode = 'binary'
)
```

```python
model2 = cnn.fit(
    train_ds,
    epochs=10,
    validation_data=test_ds)
```
